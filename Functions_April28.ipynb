```python
# meters to km
# there are 1000 meters in a km
# 2000 meters & 5 km

# formula = meters/1000 + kms
#math = 2000/1000 + 5


def convert_m(meters, kms):
    total_kms = (meters/1000) + kms
    return total_kms
```

```python
miita = int(input("Enter the meters: "))
kilomita = int(input("Enter the kilometers: "))

convert_m(miita, kilomita)
```

```python
masaa = int(input("Enter the meters: "))
dakika = int(input("Enter the kilometers: "))

convert_m(masaa, dakika)
```
